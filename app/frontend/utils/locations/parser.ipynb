```python
a = []
base_url = 'http://w1.dwar.ru/hunt_conf.php?mode=hunt_farm&area_id={}&instance_id=0'
obj_ids = ["farm_rock", "monster", "farm_floor", "farm_fish"]
```

```python
def getObj_byId():
    obj_ids = ["farm_rock", "monster", "farm_floor", "farm_fish"]
    monsters = []
    resources = []
    for _id in obj_ids:
        if _id == "monster":
            monsters = [stone for stone in soup.find_all('object', id=_id)]
        else:
            resources = [resource for resource in soup.find_all('object', id=_id)]
    return dict(monster=monsters, res=resources)
```

```python
instances = namedtuple('Instances', 'inst_name inst_id inst_url resources')
```

```python
for location in os.listdir(os.getcwd()):
    data = open(location, 'r').read() if location.endswith('xml') else False
    if data:
        soup = BeautifulSoup(data, 'html.parser')
        
        loc = soup.findAll('location')
        inst_titles = [i.title.get_text() for i in loc]
        inst_ids = [i.attrs['id'] for i in loc]
        inst_urls = [base_url.format(inst_id) for inst_id in inst_ids]
        
        x = getObj_byId()
        monster, resource = x['monster'], x['res']
#         print(inst_titles)
        _=0
        for name in loc:
            _+=1
            inst_name = name.title.get_text()
            inst_id = name.attrs['id']
            inst_url = base_url.format(inst_id)
#             print(inst_name)
            inst = instances(inst_name, inst_id, inst_url, monster, resource)
            a.append(inst)
#             print(inst_name, inst_id, inst_url)
           
                
#             if fishes:
#                 names = [c.item.get_text() for c in fishes]
                
# #                 print(names)
#                 fish_name = fishes[0].item.get_text()
#             if stones:
#                 stone_name = stones[0].item.get_text()
#             if monsters:
# #                 print(monsters)
#                 monster_name = monsters[1].item.get_text()
#                 names = [c.item.get_text() for c in monsters]
#                 print(names)
```

```python
for location in os.listdir(os.getcwd()):
    tree = et.parse(location) if location.endswith('xml') else False
    if tree:
        
        root = tree.getroot()
        for child in root.findall('area/location'):
#             print(dir(child))
            location_name = child.getchildren()[0].text
            location_id = child.attrib['id']
            location_url = base_url.format(location_id)
            all_resource = []
            for obj in child.getchildren():
                
                if obj.tag == 'object':

                    for res in obj.getchildren():

                        if res.tag == 'item':
                            all_resource.append(res.text)
            res = instances(location_name, location_id, location_url, all_resource)
            a.append(res)
```
